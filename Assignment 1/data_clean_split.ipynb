# Creating Train and Valid Splits

```python
unique_cust_ids = df_transactions["cust_id"].unique()
```

```python
train_ids, valid_ids = train_test_split(
    unique_cust_ids, 
    test_size=0.20, 
    random_state=7
)
```

```python
df_train = df_transactions[df_transactions["cust_id"].isin(train_ids)].copy()
df_valid = df_transactions[df_transactions["cust_id"].isin(valid_ids)].copy()
```

```python
df_train.to_csv("data/train_clean.csv", index=False)
```

```python
df_valid.to_csv("data/valid_clean.csv", index=False)
```

```python
df_targets = pd.read_csv("data/customer_clv_train.csv")
```

```python
df_targets.head()
```

```python
train_targets = df_targets[df_targets["cust_id"].isin(train_ids)]
```

```python
valid_targets = df_targets[df_targets["cust_id"].isin(valid_ids)]
```

```python
train_targets.to_csv("data/train_targets.csv", index=False)
```

```python
valid_targets.to_csv("data/valid_targets.csv", index=False)
```
